```python
import numpy as np
# make a 2 by 3 matrix

def normalized_row(x):
    x_norm = np.linalg.norm(x, ord=2,axis=1, keepdims=True)
    x = x / x_norm
    print(x)
    print(f"x shape: {x.shape}")
    print(f"x_norm  shape: {x_norm.shape}")
    
    return x


def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    x_exp = np.exp(x)
    x_sum = np.sum(x_exp, axis=1, keepdims=True)
    s = x_exp / x_sum
    return s


def L1(yhat, y):
    """
    Calculate L1 loss (Mean Absolute Error)

    Arguments:
    yhat -- predicted values (numpy array)
    y -- true values (numpy array)

    Returns:
    loss -- L1 loss value (scalar)
    """
    loss = np.sum(np.abs(y - yhat))
    return loss


def L2(yhat, y):
    """
    Calculate L2 loss (Mean Squared Error)

    Arguments:
    yhat -- predicted values (numpy array)
    y -- true values (numpy array)

    Returns:
    loss -- L2 loss value (scalar)
    """
    loss = np.sum((y - yhat) ** 2)
    return loss
```
